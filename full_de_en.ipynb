```python
import getpass
from langchain_openai import ChatOpenAI

# 1. Securely get your OpenRouter API key
try:
    api_key = getpass.getpass('Enter your OpenRouter API Key: ')
except Exception as e:
    print("Could not read API key. Please ensure you are running this in a TTY environment.")
    api_key = None

if api_key:
    # 2. Define the optional headers for OpenRouter using only ASCII characters
    # Do NOT include emojis or other special characters here.
    openrouter_headers = {
        "HTTP-Referer": "https://colab.research.google.com/",
        "X-Title": "Colab PDF Translator"
    }

    # 3. Configure ChatOpenAI directly, now including the default_headers
    llm = ChatOpenAI(
        model="openai/gpt-oss-20b",
        openai_api_key=api_key,
        openai_api_base="https://openrouter.ai/api/v1",
        temperature=0,
        default_headers=openrouter_headers
    )

    print("‚úÖ Configured to use 'openai/gpt-oss-20b' via OpenRouter with custom headers.")
else:
    print("‚ùóÔ∏è API key not provided. The following cells will not work.")
```

```python
from bs4 import BeautifulSoup

def translate_text(text: str) -> str:
    """Uses the API to translate text from German to English."""
    prompt = f"""You are an expert financial translator. Translate the following German text to English.
Your response must contain ONLY the English translation and nothing else.

GERMAN TEXT:
"{text}"

ENGLISH TRANSLATION:
"""
    try:
        response = llm.invoke(prompt)
        return response.content.strip()
    except Exception as e:
        print(f"üö® API Error during translation: {e}. Returning original text as a fallback.")
        return text

def grade_translation(original_text: str, translated_text: str) -> dict:
    """Uses the API to grade the English translation."""
    # CHANGE: The role is now an English proofreader.
    prompt = f"""You are a meticulous English financial proofreader. Your task is to grade a translation.
Original German Text:
<original>
{original_text}
</original>

Proposed English Translation:
<translation>
{translated_text}
</translation>

Analyze the proposed translation for accuracy, fluency, and correct financial terminology.
Then, provide your decision inside <decision> tags. The decision must be EXACTLY one of: APPROVE or REJECT.
If you reject, you MUST provide a brief, actionable critique inside <critique> tags.
"""
    response_content = ""
    try:
        response = llm.invoke(prompt)
        response_content = response.content
    except Exception as e:
        print(f"üö® API Error during grading: {e}. Defaulting to REJECT.")

    try:
        decision = response_content.split('<decision>')[1].split('</decision>')[0].strip().upper()
        if "APPROVE" in decision:
            return {"grade": "APPROVE", "critique": "None"}
        elif "REJECT" in decision:
            try:
                critique = response_content.split('<critique>')[1].split('</critique>')[0].strip()
                return {"grade": "REJECT", "critique": critique if critique else "No critique provided."}
            except IndexError:
                print(f"‚ö†Ô∏è Warning: REJECT decision was made without a critique. Output: {response_content}")
                return {"grade": "REJECT", "critique": "Model failed to provide a critique for rejection."}
        else:
            raise IndexError
    except IndexError:
        print(f"‚ö†Ô∏è Warning: Model output was malformed. Defaulting to REJECT. Output: {response_content}")
        return {"grade": "REJECT", "critique": "Model output did not contain a valid <decision> tag."}


def retranslate_with_feedback(original_text: str, failed_translation: str, critique: str) -> str:
    """Uses the API to attempt a new English translation based on feedback."""
    prompt = f"""You are an expert translator. Your previous translation was rejected.
Original German Text: "{original_text}"
Your failed English translation: "{failed_translation}"
Critique: "{critique}"
Provide a new, corrected English translation that addresses the critique. Output ONLY the corrected English text."""
    try:
        response = llm.invoke(prompt)
        return response.content.strip()
    except Exception as e:
        print(f"üö® API Error during re-translation: {e}. Returning failed translation as fallback.")
        return failed_translation

def choose_best_translation(original_text: str, translation_a: str, translation_b: str) -> str:
    """Uses the API to choose the best English translation."""
    prompt = f"""You are a final arbiter. Choose the BEST translation.
Original German Text: "{original_text}"
Translation A (The first, direct translation): "{translation_a}"
Translation B (The last, corrected attempt): "{translation_b}"
Which English translation is better, even if imperfect? Respond ONLY with the full text of the better translation."""
    try:
        response = llm.invoke(prompt)
        return response.content.strip()
    except Exception as e:
        print(f"üö® API Error during final choice: {e}. Defaulting to the last attempt.")
        return translation_b

print("‚úÖ Agent tools configured for German-to-English translation.")
```

```python
from typing import List, TypedDict
from langgraph.graph import StateGraph, END

class GraphState(TypedDict):
    original_text: str
    translation_history: List[str]
    critique_history: List[str]
    retry_count: int
    final_translation: str
    grade: str

def translate_node(state: GraphState):
    print("---NODE: TRANSLATE---")
    translation = translate_text(state['original_text'])
    return {"translation_history": [translation], "retry_count": 0}

def grade_node(state: GraphState):
    print("---NODE: GRADE---")
    grade_result = grade_translation(state['original_text'], state['translation_history'][-1])
    return {"critique_history": state['critique_history'] + [grade_result['critique']], "grade": grade_result['grade']}

def retranslate_node(state: GraphState):
    print(f"---NODE: RETRANSLATE (Attempt {state['retry_count'] + 1})---")
    corrected_translation = retranslate_with_feedback(state['original_text'], state['translation_history'][-1], state['critique_history'][-1])
    return {"translation_history": state['translation_history'] + [corrected_translation], "retry_count": state['retry_count'] + 1}

def choose_best_node(state: GraphState):
    print("---NODE: CHOOSE BEST---")
    best_translation = choose_best_translation(state['original_text'], state['translation_history'][0], state['translation_history'][-1])
    return {"final_translation": best_translation}

def finalize_node(state: GraphState):
    print("---NODE: FINALIZE---")
    return {"final_translation": state['translation_history'][-1]}

def should_continue(state: GraphState):
    print("---EDGE: SHOULD CONTINUE?---")
    if state['grade'] == "APPROVE":
        print("DECISION: Translation approved. Finalizing.")
        return "finalize"
    elif state['retry_count'] >= 2: # Limit retries to 2 for speed
        print("DECISION: Max retries reached. Choosing best option.")
        return "choose_best"
    else:
        print("DECISION: Translation rejected. Retrying.")
        return "retranslate"

# Define and compile the graph
graph = StateGraph(GraphState)
graph.add_node("translate", translate_node)
graph.add_node("grade", grade_node)
graph.add_node("retranslate", retranslate_node)
graph.add_node("choose_best", choose_best_node)
graph.add_node("finalize", finalize_node)

graph.set_entry_point("translate")
graph.add_edge("translate", "grade")
graph.add_edge("retranslate", "grade")
graph.add_edge("choose_best", END)
graph.add_edge("finalize", END)
graph.add_conditional_edges(
    "grade", should_continue, {
        "retranslate": "retranslate", "choose_best": "choose_best", "finalize": "finalize"
    }
)

agent = graph.compile()
print("‚úÖ LangGraph translation agent compiled.")
```

```python
import os
from IPython.display import display, HTML, Markdown
from unstructured.partition.pdf import partition_pdf

def translate_table(html_table: str, agent_instance):
    """Iterates through table cells and translates their content."""
    soup = BeautifulSoup(html_table, 'html.parser')
    cells = soup.find_all(['td', 'th'])
    for i, cell in enumerate(cells):
        if cell.string and cell.string.strip():
            original_text = cell.string.strip()
            print(f"    - Translating table cell {i+1}/{len(cells)}: '{original_text[:40]}...'")
            state = agent_instance.invoke({
                "original_text": original_text,
                "translation_history": [], "critique_history": [], "retry_count": 0
            })
            translated_text = state['final_translation']
            cell.string.replace_with(translated_text)
    return str(soup)

# Check if a file was uploaded in the previous cell
if 'pdf_path' in locals() and os.path.exists(pdf_path):
    print("--- Starting PDF Translation Process ---")

    # 1. Parse the document structure
    print("\nStep 1/3: Parsing document structure...")
    all_elements = partition_pdf(
        filename=pdf_path,
        strategy="hi_res",
        infer_table_structure=True,
        languages=['deu'] # to specify the source language is German
    )
    elements_to_translate = [el for el in all_elements if len(el.text) > 10]
    total_elements = len(elements_to_translate)
    print(f"PDF parsed. Found {total_elements} elements to translate.")

    # 2. Translate each element
    print("\nStep 2/3: Translating elements one by one...")
    translated_html_parts = []
    for i, element in enumerate(elements_to_translate):
        print(f"\n-> Translating element {i+1}/{total_elements} (Type: {element.category})")

        if element.category == "Table":
            html_table = getattr(element.metadata, 'text_as_html', None)
            if html_table:
                translated_table_html = translate_table(html_table, agent)
                translated_html_parts.append(translated_table_html)
        else:
            initial_state = {
                "original_text": element.text,
                "translation_history": [], "critique_history": [], "retry_count": 0,
            }
            final_state = agent.invoke(initial_state)
            final_translation = final_state['final_translation']

            if element.category in ["Title", "Header", "SubTitle"]:
                translated_html_parts.append(f"<h2>{final_translation}</h2>")
            else:
                translated_html_parts.append(f"<p>{final_translation}</p>")

    # 3. Display the final result
    print("\nStep 3/3: Assembling final document...")
    full_translation_html = "\n".join(translated_html_parts)

    display(Markdown("---"))
    display(Markdown("## üá¨üáß Final English Translation üá¨üáß"))
    display(HTML(f'''
    <div style="height: 600px; overflow-y: auto; border: 2px solid #ccc; padding: 20px;
                 border-radius: 8px; background-color: #f9f9f9; font-family: 'Arial', sans-serif;">
        {full_translation_html}
    </div>
    '''))

    # Clean up the uploaded file
    os.remove(pdf_path)

else:
    print("‚ùóÔ∏è Please run Cell 5 to upload a PDF file before running this cell.")
```
